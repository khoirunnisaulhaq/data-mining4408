```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_synthetic_data(num_samples=100, noise=100):
    np.random.seed(42)
    X = np.random.rand(num_samples, 1) * 100  # Luas Tanah (misalnya, dari 0 hingga 100)
    y = 2.5 * X + np.random.randn(num_samples, 1) * noise  # Harga Kontrakan (2.5 adalah koefisien regresi dan tambahan noise)
    return X, y
```

```python
X, y = generate_synthetic_data(num_samples=100, noise=100)
```

```python
plt.scatter(X, y)
plt.xlabel('Luas Tanah')
plt.ylabel('Harga Kontrakan')
plt.title('Dataset Harga Kontrakan Berdasarkan Luas Tanah')
plt.show()
```
